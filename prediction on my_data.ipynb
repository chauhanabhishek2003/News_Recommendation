```python
import numpy as np
import pandas as pd
news= pd.read_csv('my_data.csv')
```

```python
news.dropna(inplace=True)
```

```python
news['Description']=news['Description'].apply(lambda x:x.split())
news['Title2']=news['Title'].apply(lambda x:x.split())
news['Keywords']=news['Keywords'].apply(lambda x:x.split())
```

```python
news['tags']=news['Description']+news['Keywords']+news['Title2']
```

```python
df=news[['Title','URL','tags']]
df.reset_index(drop=True, inplace=True) 
```

```python
df
```

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import re
ps=PorterStemmer()
stop_words = set(stopwords.words('english'))
```

```python
df['tags']=df['tags'].apply(lambda x:" ".join(x))
df['tags']=df['tags'].apply(lambda x:x.lower())
df['tags']=df['tags'].apply(lambda x:re.sub(r'[^a-zA-Z\s]',"",x))
df['tags']=df['tags'].apply(lambda x: ' '.join([ps.stem(word) for word in x.split() if word not in stop_words]))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
bow=cv.fit_transform(df['tags']).toarray()
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sim=cosine_similarity(bow)
sim[0].shape
```

```python
def recommend(newss):
    news_idx= df[df['Title']== newss].index[0]
    lit=sorted(list(zip(df['Title'].tolist(),sim[news_idx].tolist())),reverse=True,key=lambda x:x[1])[0:6]
    for i in lit:
        x=df[df['Title']==i[0]].index[0]
        print(df.iloc[x].Title+"\n")
        if(x==news_idx):
            print("\nRECOMMENDATIONS:")
```

```python
recommend('The sweet music of jack wood')
```
